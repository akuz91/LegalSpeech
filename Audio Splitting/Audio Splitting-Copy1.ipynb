```python
def getMeta(docket, data):
    
    #get meta data as well as rearrange to desirable formal
    transcript, speakers, speaker_roles, times = data[docket]
    
    times_new = []
    for t in times:
        flatten = [item for sublist in t for item in sublist]
        times_new.append(flatten)
    
    speaker_roles = [item for sublist in speaker_roles for item in sublist]
    
    speakers =[re.sub(r"[^\w\s]", '', s) for s in speakers]
    # Replace all runs of whitespace with underscore
    speakers =[re.sub(r"\s+", '_', s) for s in speakers]
    
    return transcript, speakers, speaker_roles, times_new
```

```python
transcript, speakers, speaker_roles, times = getMeta('17-1268',data)
```

```python
def getSpeakerDict(transcript, speakers, speaker_roles, times):
    speaker_dict = {}
    for i in range(len(speakers)):
        speaker_dict[speakers[i] + '_' + speaker_roles[i] + '_' + str(i)] = times[i][0], times[i][-1], transcript[i]
        
    return speaker_dict
```

# Directory Commands #
    (1) For each docket create new folder (maybe parse through the file of downloaded wav files)
    (2) For each speaker create sub folder
    (3) For each audio file, split audio according to speaker and store in speaker folder

```python
def main_script(file_path = '/oyez_metadata_11_to_20.json'):
    
    with open(os.getcwd() + file_path) as f:
        data = json.load(f)
    
    for docket in data:

        #get meta data
        transcript, speakers, speaker_roles, times = getMeta(docket,data)

        #create speaker dict
        speaker_dict = getSpeakerDict(transcript, speakers, speaker_roles, times)

        #create folder for docket and then sub folders (speaker + speaker_role) for each speaker in docket 
        createFolders(docket, speakers, speaker_roles, times, data)

        #split and move to correct folder 
        getSplittingCommands(docket, speaker_dict)

        break
```

```python
main_script()
```
